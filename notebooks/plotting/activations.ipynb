```python
import numpy as np
import matplotlib.pyplot as plt
import torch as th
import seaborn as sns
import os
```

```python
x = th.linspace(-3.5,3.5,101)

y_sigmoid = th.sigmoid(x)
y_tanh = th.tanh(x)
y_relu = th.relu(x)
y_lrelu = th.nn.LeakyReLU()(x)
```

```python
fig, ax = plt.subplots(figsize = (5,2.5), constrained_layout=True)

ax.axis('equal')
colors = sns.color_palette()

ax.plot(x, y_sigmoid, color=colors[0], label='Sigmoid')
ax.plot(x, y_tanh, color=colors[1], label='Tanh')
ax.plot(x, y_relu, color=colors[2], label='ReLU')
ax.plot(x, y_lrelu, color=colors[3], linestyle='--', label='Leaky ReLU')

ax.set_xlim([-3, 3])
ax.set_ylim([-1, 2])

ax.set_yticks([-1,0,1,2])

ax.set_xlabel(r'$a$')
ax.set_ylabel(r'$f(a)$')

ax.legend()

# save fig
fig_path = '/home/leonriccius/Documents/Leon_MastersThesis/figures_thesis/machine_learning'
fig_name = os.sep.join([fig_path, 'activation_functions.pdf'])
# plt.savefig(fig_name, format='pdf')
```
